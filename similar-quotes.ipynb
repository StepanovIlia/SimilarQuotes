# **Using Cosine Similarity To Group Similar Quotes**

```python
import numpy as np
import spacy

#need for vectors
nlp = spacy.load('en_core_web_lg')
```

We create word vectors for each quote. We will then compare word vectors to determine semantically similar quotes.

```python
file = open('../input/popular-quotes/goodreads_popular_quotes.txt', 'r')

#keep track of each quote
data = []
#keep track of each word vector
vectors = []

for line in file:
    data.append(line)
    vectors.append(nlp(line).vector)
  
print(len(data))
```

Cosine similarity is a method used to compare the likeness of two documents. Given two vectors *a* and *b*, we find the cosine of the angle between them.

```python
def cosine_similarity(a, b):
    return a.dot(b)/np.sqrt(a.dot(a) * b.dot(b))
```

```python
#2D array to document the relationship between each quote
similarities = [[x for x in range(len(vectors))] for y in range(len(vectors))]

for v in range(len(vectors)-1):
    for n in range(v+1, len(vectors)):
        #cosine similarity for each pair of quotes
        similarities[v][n] = cosine_similarity(vectors[v], vectors[n])
        similarities[n][v] = similarities[v][n]
```

```python
from random import randint

random_quote = randint(0,len(data)-1)
print(data[random_quote])

print("Similar Quotes:")
#order similar quotes from "most similar" to "least similar"
similar_quotes = sorted(similarities[random_quote], reverse=True)

count = 0

for i in range(len(similar_quotes)):
    #remove noise, print only 5 quotes
    if similar_quotes[i] < 1 and count < 5:
        if data[random_quote] != data[i]:
            print(data[i])
            count += 1
```
